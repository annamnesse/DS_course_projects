Подберем признаки для обучения модели

```python
# Функция для нахождения скользящего среднего
def make_features(data, column, max_lag, rolling_mean_size):
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['month'] = data.index.month
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data[column].shift(lag)

    data['rolling_mean'] = data[column].shift().rolling(rolling_mean_size).mean()
```

<div class="alert alert-block alert-info">
<b>Совет: </b> Желательно было написать функцию таким образом, чтобы она возвращала новый датафрейм, а не изменяла глобальную переменную.
</div>

```python
# Подберем показатели максимального лага и скользящего среднего
max_lag = 25
rolling_mean_size = 35
```

```python
# Признаки для обучения модели
make_features(taxi, 'num_orders', max_lag, rolling_mean_size)

train, test = train_test_split(taxi, shuffle=False, test_size=0.1)
train = train.dropna()

train_features = train.drop('num_orders', axis=1)
train_target = train['num_orders']
test_features = test.drop('num_orders', axis=1)
test_target = test['num_orders']
```

```python
#Размеры обучающей и тестовой выборки
print('Размер обучающей выборки: {} объектов, {} признаков'.format(train_features.shape[0], train_features.shape[1]))
print('Размер тестовой выборки: {} объектов, {} признаков'.format(test_features.shape[0], test_features.shape[1]))
```

<div class="alert alert-block alert-success">
<b>Успех:</b> Разбиение было сделано верно.
</div>

Промасштабируем количественные признаки обучающей и тестовой выборок ввиду заметных выбросов.

```python
numeric = ['rolling_mean'] + train_features.columns[train_features.columns.str.contains('lag')].to_list()
scaler = StandardScaler()
scaler.fit(train_features[numeric])

train_features[numeric] = scaler.transform(train_features[numeric])
test_features[numeric] = scaler.transform(test_features[numeric])
```

<div class="alert alert-block alert-success">
<b>Успех:</b> Отлично, что scaler был обучен только на треине.
</div>

```python
# Функция расчета RMSE
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

Обучим модель линейной регрессии

```python
model_lr = LinearRegression()
model_lr.fit(train_features, train_target)
train_predicted_lr = model_lr.predict(train_features)
test_predicted_lr = model_lr.predict(test_features)

print("RMSE на обучающей выборке: ", round(rmse(train_predicted_lr, train_target), 3))
print("RMSE на тестовой выборке: ", round(rmse(test_predicted_lr, test_target), 3))
```

Модель линейной регрессии близка по качеству заданному условию анализа: на тестовой выборке RMSE составляет **48.673**. Начнем поиск идеальной модели.

Проведем обучение моделей, получим предсказания и сравним их с тестовыми данными.

```python
# Модель линейной регрессии
model = LinearRegression()
model.fit(train_features, train_target)
target_predict = model.predict(test_features)
final_RMSE_lr = mean_squared_error(test_target, target_predict) ** 0.5

plt.figure(figsize=(15, 7))
plt.plot(test_target.index, 
         test_target, 
         'b-', 
         test_target.index, 
         target_predict,
         'r-')
```

```python
# Модель случайного леса
model = RandomForestRegressor(n_estimators = 81, max_depth=9, random_state=12345)
model.fit(train_features, train_target)
target_predict = model.predict(test_features)
final_RMSE_rfr = mean_squared_error(test_target, target_predict) ** 0.5

plt.figure(figsize=(15, 7))
plt.plot(test_target.index, 
         test_target, 
         'b-',
         test_target.index, 
         target_predict,
         'r-')
```

```python
# Модель решающего дерева
model = DecisionTreeRegressor(max_depth = 6, random_state=12345)
model.fit(train_features, train_target)
target_predict = model.predict(test_features)
final_RMSE_dtr = mean_squared_error(test_target, target_predict) ** 0.5

plt.figure(figsize=(15, 7))
plt.plot(test_target.index, 
         test_target, 
         'b-', 
         test_target.index, 
         target_predict,
         'r-')
```

```python
# Модель CatBoost
model = CatBoostRegressor(iterations = 100, learning_rate = 0.2, random_state=12345, verbose=False)
model.fit(train_features, train_target)
target_predict = model.predict(test_features)
final_RMSE_cbr = mean_squared_error(test_target, target_predict) ** 0.5

plt.figure(figsize=(15, 7))
plt.plot(test_target.index, 
         test_target, 
         'b-', 
         test_target.index, 
         target_predict,
         'r-')
```

```python
# Модель LightGBM
model = LGBMRegressor(learning_rate = 0.1, random_state=12345)
model.fit(train_features, train_target)
target_predict = model.predict(test_features)
final_RMSE_lgbm = mean_squared_error(test_target, target_predict) ** 0.5

plt.figure(figsize=(15, 7))
plt.plot(test_target.index, 
         test_target, 
         'b-', 
         test_target.index, 
         target_predict,
         'r-')
```
