```python
data['ceiling_height'].isna().mean()
```

Сбрасывать со счетов более трети пропусков нецелесообразно. Будем пытаться заполнить адекватной заменой.
Для начала проанализируем эту выборку по значениям.

```python
data['ceiling_height'].describe()
```

В целом данные выглядят адекватно, за исключением подозрительного выброса в 1 и 100 метров. Пропуск данных, как и странные выбросы минимума и максимума могли быть связаны просто с незнанием точного значения высоты потолка заполнявшим.

Для заполнения пропусков я бы исключила минимальное и максимальное значение и взяла бы среднюю высоту потолка. Поскольку 100 - это явный вылет, исключим экстремально высокие потолки из выборки.

```python
data.drop(data[data['ceiling_height'] > 5].index, inplace=True)
```

```python
data.boxplot(column = 'ceiling_height')
plt.show()
```

Пока не готова анализировать глубже, но считаю целесообразным заполнить пропуски медианными значениями, так как довольно много вылетов выше "усов".

```python
data['ceiling_height']=data['ceiling_height'].fillna(data['ceiling_height'].median())
```

#### Этажность дома floors_total

```python
data['floors_total'].isna().mean()
```

Менее 0.5 % пропусков - не так много. Все-таки займемся судьбой пропущенных значений и изучим корреляционные связи данного столбца.

```python
data.corr().sort_values(by='floors_total', ascending=False)['floors_total'].reset_index().head()
```

Очевидна более сильная связь с этажом квартиры. Значит, попробуем провести заполнение с учетом указанного этажа квартиры.

Создадим словарь с указанным этажом квартиры и соответствующим ему медианным значением этажности дома.

```python
floor_dict = data.pivot_table(index='floor', values='floors_total', aggfunc='median').reset_index()
print(floor_dict)
```

Прогоним в цикле заполнение пропусков этажности дома медианным значением по указанному этажу квартиры из созданного словаря floor_dict

```python
for floor in range(data['floor'].min(), (data['floor'].max()+1)):
    data.loc[data['floor'] == floor, 'floors_total'] = (
    data.loc[data['floor'] == floor, 'floors_total']
    .fillna(floor_dict[floor_dict['floor'] == floor]['floors_total'].sum()))
```

```python
data['floors_total'].describe()
```

```python
living_dict = data.pivot_table(index='rooms', values='living_ratio', aggfunc='median').reset_index()
living_dict
```

Добавим новый столбец floor_category с 3 категориями по этажу: первый, последний и другой.

Функцию введем по следующей логике: первый этаж, если значение в столбце floor равно 1; последний этаж, если значение равно этажности дома (floors_total); в противном случае - другой этаж. 

```python
def floor_marker(row):
        if row[8] == 1:
            return 'Первый этаж'
        elif row[8] == row[6]:
            return 'Последний этаж'
        else:
            return 'Другой этаж'
```

#### Зависимость цены квартиры от дня недели размещения

Извлечем для дальнейшейго анализа из столбца first_day_exposition отдельные столбцы с днем недели, месяцем и годом.

```python
data['weekday_exposition']=data['first_day_exposition'].dt.weekday.astype('str').map({'0': 'пн', '1': 'вт', '2': 'ср', '3': 'чт', '4': 'пт', '5': 'сб', '6': 'вс'})
data['month_exposition']=data['first_day_exposition'].dt.month.astype('str').map({'1': 'янв', '2': 'фев', '3': 'мар', '4': 'апр', '5': 'май', '6': 'июн', '7': 'июл', '8': 'авг', '9': 'сен', '10': 'окт', '11': 'ноя', '12': 'дек'})
data['year_exposition']=data['first_day_exposition'].dt.year
```

```python
weekday_exposition=data.pivot_table(index='weekday_exposition', values='last_price', aggfunc='mean').reset_index()
days = ['пн','вт','ср','чт','пт','сб','вс']
mapping = {day: i for i, day in enumerate(days, 1)}
key = weekday_exposition['weekday_exposition'].map(mapping)
weekday_exposition = weekday_exposition.iloc[key.argsort()].set_index('weekday_exposition')
weekday_exposition.plot(kind='bar', figsize=(15, 4),rot=360)
plt.ylim(bottom=4)
plt.title('Зависимость цены от дня размещения объявления', size=14)
plt.xlabel('Дни недели')
plt.ylabel('Цена квартиры, млн.руб.')
plt.show()
```

<div class="alert alert-block alert-info">
При рассмотрении зависимости средней цены квартир по дням размещения объявления на неделе видна более сглаженная картина с плавным ростом к середине недели (вторник, среда, четверг) и дальнейшим спадом к выходным дням. </div>

#### Зависимость цены от месяца размещения

```python
month_exposition = data.pivot_table(index='month_exposition', values='last_price', aggfunc='mean').reset_index()
months = ['','янв', 'фев', 'мар', 'апр', 'май', 'июн', 'июл', 'авг', 'сен', 'окт', 'ноя', 'дек']
mapping = {month: i for i, month in enumerate(months)}
key = month_exposition['month_exposition'].map(mapping)
month_exposition = month_exposition.iloc[key.argsort()].set_index('month_exposition')
month_exposition.plot(kind='bar', figsize=(15, 4),rot=360)
plt.ylim(bottom=4)
plt.title('Зависимость от месяца размещения объявления', size=14)
plt.xlabel('Месяцы')
plt.ylabel('Средняя цена квартир, млн.руб.')
plt.show()
```
